```python
# Converting 'Date' column to datetime
sales_data['Date'] = pd.to_datetime(sales_data['Date'])
```

```python
sales_data=df_sales.copy()
```

```python

# Overall Sales Summary
total_revenue = sales_data['Total'].sum()
average_sales = sales_data['Total'].mean()
total_transactions = sales_data.shape[0]
```

```python

# Sales by Branch
sales_by_branch = sales_data.groupby('Branch')['Total'].agg(['sum', 'mean', 'count'])

# Sales by Product Line
sales_by_product_line = sales_data.groupby('Product line')['Total'].agg(['sum', 'mean', 'count'])

# Customer Demographics
sales_by_customer_type = sales_data.groupby('Customer type')['Total'].agg(['sum', 'mean', 'count'])
sales_by_gender = sales_data.groupby('Gender')['Total'].agg(['sum', 'mean', 'count'])

# Sales over Time
sales_over_time = sales_data.groupby(sales_data['Date'].dt.month)['Total'].agg(['sum', 'mean', 'count'])

# Payment Method Usage
sales_by_payment_method = sales_data.groupby('Payment')['Total'].agg(['sum', 'mean', 'count'])
```

```python
# Print the results
print("Overall Sales Summary:")
print("Total Revenue:", total_revenue)
print("Average Sales:", average_sales)
print("Total Transactions:", total_transactions)

print("\nSales by Branch:")
print(sales_by_branch)

print("\nSales by Product Line:")
print(sales_by_product_line)

print("\nCustomer Demographics:")
print("Sales by Customer Type:")
print(sales_by_customer_type)
print("Sales by Gender:")
print(sales_by_gender)

print("\nSales over Time:")
print(sales_over_time)

print("\nPayment Method Usage:")
print(sales_by_payment_method)
```
